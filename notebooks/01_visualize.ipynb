```python
import sys
import os

sys.path.append("..")

import torch
import matplotlib.pyplot as plt

from benchmarks.functions import easy_four_gaussian

print("Current working directory:", os.getcwd())


# ======================================================
# Visualization
# ======================================================

n_grid = 300

x1 = torch.linspace(0, 1, n_grid)
x2 = torch.linspace(0, 1, n_grid)

X1, X2 = torch.meshgrid(x1, x2)

X = torch.stack([
    X1.reshape(-1),
    X2.reshape(-1)
], dim=1)

Z = easy_four_gaussian(X).reshape(n_grid, n_grid).cpu()

plt.figure(figsize=(8, 6))
plt.contourf(
    X1.numpy(),
    X2.numpy(),
    Z.numpy(),
    levels=50
)

plt.colorbar()
plt.title("Easy Four Gaussian Benchmark")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tight_layout()
plt.savefig("/direct/usatlas+u/zhaozhong/nf_integrator/benchmarks/plots/easy_four_gaussian.png", dpi=300)
plt.show()
```
